<h3>üìö  <font color='lightpink'>1. Els≈ë l√©p√©s: A sz√ºks√©ges programoz√°si k√∂nyvt√°rak import√°l√°sa</font></h3>


```python
import tensorflow as tf
from keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np
```

<h3>üíΩ <font color='lightpink'>2. l√©p√©s - A tan√≠t√°si adatok - adat√°llom√°ny bet√∂lt√©se</font></h3>


*   A g√©pi tanul√°shoz rengeteg adatra, jelen esetben k√©pre van sz√ºks√©g.<br>
*   A toront√≥i egyetem j√≥volt√°b√≥l l√©tezik a "CIFAR10" n√©vre hallgat√≥ k√©p√°llom√°ny.<br>

Benne 60,000 felv√©tel. Mivel ez egy tanul√°sra el≈ëk√©sz√≠tett adat, ez√©rt egy sor k√≥ddal ezt be tudjuk t√∂lteni.

```python
(X_train, y_train), (X_test,y_test) = datasets.cifar10.load_data()
```

<h3><font color='lightpink'>3. A neur√°lis h√°l√≥zat (NN) fel√©p√≠t√©se tanul√°shoz</font></h3>
Al√°bbi k√≥dban meghat√°rozzuk az NN r√©tegeit √©s a tanul√°si folyamat f≈ëbb param√©tereit.

```python
X_train = X_train / 255.0
X_test = X_test / 255.0

classes = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]

cnn = models.Sequential([
    layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

cnn.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = cnn.fit(X_train, y_train, epochs=8, validation_data=(X_test, y_test))
print("Lefutott a tanul√°s mehetsz tov√°bb")
# FONTOS >> A tanitas lefutasat meg kell varni ez tobb percet vesz igenybe
```

<h3><font color='lightcyan'>Fontos!! Tesztel√©s el≈ëtt v√°rd meg a</font><br> "Lefutott a tanul√°s mehetsz tov√°bb" ki√≠r√°st</h3>

<h1><font color='lightblue'>Tesztel√©s</font></h1>
<h3><font color='lightpink'>4. A feltan√≠tott neur√°lis h√°l√≥ tesztel√©se</font></h3>
Itt els≈ëre a feltan√≠t√°shoz is haszn√°lt 'cifar10' adatset, teszt r√©sz√©ben szerepl≈ë k√©pek k√∂z√ºl v√°lasztunk ki p√°rat random.<br> Majd megk√©rj√ºk a frissen feltan√≠tott modellt, hogy √°llap√≠tsa meg, hogy mi tal√°lhat√≥ a k√©pen?!
<p><font color='lightpink'>Itt is csak nyomd meg a play gombot, a kis tesztk√©pek meg fognak jelenni a k√≥d alatt!</font></p>

```python
def display_images(images, labels, predictions, classes, num_images=5):
    plt.figure(figsize=(8, 2))
    for i in range(num_images):
        plt.subplot(1, num_images, i + 1)
        plt.imshow(images[i])
        plt.title(f"True: {classes[labels[i][0]]}\nPredicted: {classes[predictions[i]]}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# P√°r k√©p random kiv√°laszt√°sa a teszt adat√°llom√°nyb√≥l
num_images_to_display = 5
random_indices = np.random.choice(len(X_test), num_images_to_display, replace=False)
images_to_display = X_test[random_indices]
labels_to_display = y_test[random_indices]

predictions = np.argmax(cnn.predict(images_to_display), axis=-1)

display_images(images_to_display, labels_to_display, predictions, classes)
```

<h3><font color='lightpink'>5.2 Az el≈ëbbi l√©p√©sben felt√∂lt√∂tt k√©peket tesztelheted itt</font></h3>

```python
from keras.preprocessing.image import load_img, img_to_array
import io
import os

# A fenti l√©p√©sben let√∂lt√∂tt teszt mapp√°t itt √°t tudod √≠rni ha szeretn√©d
test_folder = '/content/test1'  #

num_images_to_display = len(os.listdir(test_folder))  # Number of images in the 'test1' folder
plt.figure(figsize=(10, 3))

for idx, filename in enumerate(os.listdir(test_folder), start=1):
    file_path = os.path.join(test_folder, filename)
    plt.subplot(2, num_images_to_display, idx)

    orig_img = load_img(file_path)
    img = load_img(file_path, target_size=(32, 32))  # Target size
    img_array = img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    processed_img = img_array / 255.0  # Normalization

    prediction = cnn.predict(processed_img)
    predicted_class = np.argmax(prediction)

    plt.imshow(orig_img)
    plt.axis('off')
    predicted_label = classes[predicted_class]

    text_y = orig_img.size[1] + 90
    plt.text(10, text_y, f"{predicted_label}", fontsize=16)

plt.show()
```
